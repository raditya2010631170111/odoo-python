3) Menghapus fitur konstan

```python
from sklearn.feature_selection import VarianceThreshold
var_threshold = VarianceThreshold(threshold = 0) 
var_threshold.fit(X)
var_threshold.get_support()
```

4) Menghapus fitur quasi-konstan

```python
var_threshold = VarianceThreshold(threshold = 0.1) 
var_threshold.fit(X)
var_threshold.get_support()
```

True = high variance. Karena semua fitur sudah high variance maka tidak diperlukan proses mengeluarkan fitur low variance
