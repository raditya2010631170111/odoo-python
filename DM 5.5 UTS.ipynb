e. Lakukan seleksi fitur pada data tersebut! (poin 15)

Seleksi Fitur dengan teknik Metode Filter memakai metode Variance threshold

```python
df.shape
```

Menghapus fitur ID, Axis=Column, X=predictor=dataku1, Y=target=t

```python
X = df.drop(['Gender', 'AttendenceEvent'],axis=1)
Y = df[['AttendenceEvent']]
Y.tail()
```

Semua fitur kategorikal telah diubah, maka tidak ada fitur numerik untuk diencode

```python
X.dtypes
```

Menghapus fitur konstan

```python
from sklearn.feature_selection import VarianceThreshold
var_threshold = VarianceThreshold(threshold = 0) 
var_threshold.fit(X)
var_threshold.get_support()
```

Menghapus fitur quasi-konstan

```python
var_threshold = VarianceThreshold(threshold = 0.1) 
var_threshold.fit(X)
var_threshold.get_support()
```
