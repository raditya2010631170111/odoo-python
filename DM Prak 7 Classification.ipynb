5. Algoritma Naive Bayes

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/raditya2010631170111/df_20111/main/heart.csv')
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
```

```python
data['Age'] = enc.fit_transform(data['Age'].values)
data['Sex'] = enc.fit_transform(data['Sex'].values)
data['ChestPainType'] = enc.fit_transform(data['ChestPainType'].values)
data['RestingBP'] = enc.fit_transform(data['RestingBP'].values)
data['Cholesterol'] = enc.fit_transform(data['Cholesterol'].values)
data['FastingBS'] = enc.fit_transform(data['FastingBS'].values)
data['RestingECG'] = enc.fit_transform(data['RestingECG'].values)
data['MaxHR'] = enc.fit_transform(data['MaxHR'].values)
data['ExerciseAngina'] = enc.fit_transform(data['ExerciseAngina'].values)
data['Oldpeak'] = enc.fit_transform(data['Oldpeak'].values)
data['ST_Slope'] = enc.fit_transform(data['ST_Slope'].values)
data['HeartDisease'] = enc.fit_transform(data['HeartDisease'].values)
```

```python
atr_data = data.drop(columns='HeartDisease')
atr_data.head()
```

```python
cls_data = data['HeartDisease']
cls_data.head(100)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
x_train, x_test, y_train, y_test = train_test_split(atr_data,cls_data, test_size=0.2, random_state=5000)
nb = GaussianNB()
nb.fit(x_train, y_train)
```

```python
y_pred = nb.predict(x_test)
y_pred
```

```python
np.array(y_test)
```

```python
nb.predict_proba(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
