```python
from sklearn.feature_selection import VarianceThreshold
var_threshold = VarianceThreshold(threshold = 0) #Menghapus fitur konstan
var_threshold.fit(X)
var_threshold.get_support()
```

```python
var_threshold = VarianceThreshold(threshold = 0.1) #Menghapus fitur quasi-konstan
var_threshold.fit(X)
var_threshold.get_support()
```

versi panjang

```python
low_var_cols_another = []

for column in X.columns:
    if column X.columns[var_threshold.get_support()]:
      low_var_cols_another.append(column)

for col in low_var_cols_another:
    print(col, end=', ')
```

versi pendek

```python
low_var_cols = [column for column in X.columns
         if column not in X.columns[var_threshold.get_support()]]
for col in low_var_cols:
    print(col)
```

```python
X_high = X.drop(low_var_cols, axis=1)
```

```python
X_high.sample(5)
```

```python
var_threshold.fit(X_high)
var_threshold.get_support()
```
