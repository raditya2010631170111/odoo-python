## Chi-square

```python
from sklearn.feature_selection import chi2
```

### Load Dataset

```python
df_bank = pd.read_csv('https://raw.githubusercontent.com/ammfat/datasets/main/BankChurners.csv')

df_bank.tail()
```

```python
df_bank.shape
```

### Drop Unrequired Columns

```python
df_bank.columns
```

```python
df_bank = df_bank.drop([
    'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1'
    , 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'
], axis=1)

df_bank.columns
```

### Identify Missing Values

```python
df_bank.isnull().sum()
```

### Drop Unrequired Columns

```python
df_bank = df_bank.drop(['CLIENTNUM'], axis=1)
```

### Data Transformation

```python
df_bank.info()
```

```python
df_bank.dtypes
```

```python
df_bank['Gender'].dtype
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
```

```python
# for col in df_bank.columns:
#     if df_bank[col].dtype == '0':
#         df_bank[col] = encoder.fit_transform(df_bank[col])
```

```python
df_bank['Attrition_Flag'] = encoder.fit_transform(df_bank['Attrition_Flag'])
df_bank['Gender'] = encoder.fit_transform(df_bank['Gender'])
df_bank['Education_Level'] = encoder.fit_transform(df_bank['Education_Level'])
df_bank['Marital_Status'] = encoder.fit_transform(df_bank['Marital_Status'])
df_bank['Income_Category'] = encoder.fit_transform(df_bank['Income_Category'])
df_bank['Card_Category'] = encoder.fit_transform(df_bank['Card_Category'])
```

```python
df_bank.tail()
```

### Chi-square test

```python
X = df_bank.drop(["Attrition_Flag"], axis=1)
y = df_bank['Attrition_Flag']
```

```python
chi_score = chi2(X, y)

chi_score
```

```python
p_values = pd.Series(chi_score[1],index = X.columns)
p_values.sort_values(ascending = False , inplace = True)
```

```python
p_values.plot.bar()

plt.show()
```
