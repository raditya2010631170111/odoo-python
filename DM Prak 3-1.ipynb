One-hot Encoder

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
```

```python
df['Sex'].tail()
```

```python
df.info()
```

mengubah dtype ['Sex'] object -> category & diisi [female, male]

```python
df['Sex'] = df['Sex'].astype('Category')
df['Sex'] = df['Sex'].cat.codes
```

dibawah = apabila Fe 0.0 = Male 1.0, Fe 1.0 = Male 0.0

```python
pd.DataFrame(encoder.fit_transform(df[['Sex']]).toarray())
```

```python
encoder_data = pd.DataFrame(encoder.fit_transform(df[['Sex']]).toarray())
encoder_data.columns = ['female', 'male']
```

```python
df = df.join(encoder_data)
```
